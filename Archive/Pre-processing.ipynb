```python
import pandas as pd
```

```python
data = pd.read_csv('CatS_CHEMBL.csv')
data.head()
```

```python
#_Create new dataframe with useful columns

useful_data = {'smiles': data['canonical_smiles'], 'IC50': data['standard_value'], 
               'pchembl': data['pchembl_value'], 'type': data['standard_type'], 
               'ID' : data['molecule_chembl_id'] }

data_new  = pd.DataFrame(useful_data)
data_new.head()
```

```python
#_Remove all rows that doesn't correspond to IC50

cleaned_data = data_new[data_new['type'] == 'IC50']
cleaned_data = cleaned_data.reset_index(drop = True)
cleaned_data = cleaned_data.sort_values(by=['ID'])
cleaned_data.dropna(inplace=True)
cleaned_data.head()
```

```python
cleaned_data.shape
```

```python
#_Remove duplicates using CHEMBL ID

cleaned_data.drop_duplicates(subset = 'ID', keep = 'first', inplace = True)
cleaned_data.shape
```

```python
cleaned_data.to_csv('cleaned_CatS_CHEMBL.csv')
```
